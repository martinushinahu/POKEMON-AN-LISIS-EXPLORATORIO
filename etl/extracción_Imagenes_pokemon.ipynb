AHORA CON EL APOYO DE LA POKEMON API VAMOS A DESCARGAR LAS IMAGENES DE LOS POKEMON DESDE LA 1ER GENERACIÓN HASTA LA 7TA GENERACIÓN

```python
# Importamos las librería necesarias
import requests
import pandas as pd
from io import BytesIO
from PIL import Image
import os
```

```python
# Obtenemos la información de los Pokémon desde la API
api_url = "https://pokeapi.co/api/v2/pokemon/"
num_pokemon = 800  # Número total de Pokémon
pokemon_data = []
```

```python
for i in range(1, num_pokemon + 1):
    response = requests.get(api_url + str(i))
    data = response.json()
    pokemon_data.append({
        "id": i,
        "name": data["name"],
        "image_url": data["sprites"]["front_default"]
    })
```

```python
# Creamos un DataFrame de Pandas con la información
df = pd.DataFrame(pokemon_data)
```

```python
# Creamos el directorio 'images' si no existe
os.makedirs("images", exist_ok=True)

# Descarga las imágenes y guárdalas localmente
for index, row in df.iterrows():
    response = requests.get(row["image_url"])
    img = Image.open(BytesIO(response.content))
    img.save(f"images/{row['name']}.png")
```

```python
# Agregamos la columna de ruta de la imagen al DataFrame
df["image_path"] = df["name"].apply(lambda x: f"images/{x}.png")
```

```python
# Guardamos el DataFrame en un archivo CSV
df.to_csv("pokemon_dataset_images.csv", index=False)
```
