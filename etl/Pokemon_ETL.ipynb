#### 9.1 LA MAYORÍA DE LOS POKPÉMON TIENEN CONTRAATAQUES CERCANOS A 1 EN PROMEDIO PARA LA MAYORÍA DE LOS TIPOS, LA VARIABILIDAD DE ESTOS VALORES SUGIERE QUE ALGUNOS POKÉMON PUEDEN SER PARTICULARMENTE FUERTES O DÉBILES  CONTRA CIERTOS TIPOS. ASÍ QUE VAMOS A REALIZAR UNA LISTA Y UNOS HISTOGRAMAS PARA PODER VER ESTAS INTERPRETACIONES DE ESTOS RESULTADOS.

```python
# Definimos la lista de tipos de contraataque
tipos_contraataque = ['contra_bicho', 'contra_siniestro', 'contra_dragón', 'contra_eléctrico',
                      'contra_hada', 'contra_lucha', 'contra_fuego', 'contra_volador',
                      'contra_fantasma', 'contra_planta', 'contra_tierra', 'contra_hielo',
                      'contra_normal', 'contra_veneno', 'contra_psíquico', 'contra_roca',
                      'contra_acero', 'contra_agua']
```

```python
# Identificamos los Pokémon con contraataques más altos y bajos para cada tipo
for tipo in tipos_contraataque:
    # Ordenamos de manera descendente para contraataques altos
    altos = df.sort_values(by=tipo, ascending=False)[['nombre', tipo]].head(5)

    # Ordenamos de manera ascendente para contraataques bajos
    bajos = df.sort_values(by=tipo, ascending=True)[['nombre', tipo]].head(5)

    print(f"\nContraataques más altos para {tipo}:\n{altos}")
    print(f"\nContraataques más bajos para {tipo}:\n{bajos}")
```

```python
# Visualizamos distribuciones con histogramas
plt.figure(figsize=(12, 8))
for tipo in tipos_contraataque:
    plt.subplot(4, 5, tipos_contraataque.index(tipo) + 1)
    sns.histplot(df[tipo], kde=True)
    plt.title(tipo)
plt.tight_layout()
plt.show()
```

```python
# Calculamos y visualizamos la matriz de correlación de los contraataques
matriz_correlacion = df[tipos_contraataque].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(matriz_correlacion, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Matriz de Correlación de Contraataques')
plt.show()
```

##### 10.1 EL ANÁLISIS DE LOS CONTRAATAQUES PROPORCIONADOS NOS BRINDA INFORMACIÓN SOBRE LA DIVERSIDAD Y CARACTERÍSTICAS DE LOS POKÉMON EN RELACIÓN CON SUS TIPOS Y DEBILIDADES. ESTOS RESULTADOS PUEDE SER ÚTIL PARA ESTRATEGIAS EN EL JUEGO POKEMON, YA QUE LOS ENTRENADORES PUEDES APROVECHAR LAS FORTALEZAS Y DEBILIDADES DE SUS POKÉMON EN BATALLAS.

```python
# Creamos una función para determinar si un Pokémon tiene evolución
def determinar_evolucion(pokemon, evoluciones_dict):
    if pokemon in evoluciones_dict:
        return evoluciones_dict[pokemon]
    else:
        return 'No evoluciona'
```

```python
# Aplicamo la función a la columna 'nombre' para crear la nueva columna 'Evolución'
df['Evolución'] = df['nombre'].apply(lambda x: determinar_evolucion(x, evoluciones_dict))
```

```python
# Función para calcular el daño entre dos Pokémon
def calcular_daño(pokemon1, pokemon2, ataque):
    fortalezas_pokemon2 = df.loc[df['nombre'] == pokemon2, [f'fortaleza_{tipo}' for tipo in df.columns if tipo.startswith('fortaleza_')]]
    debilidades_pokemon2 = df.loc[df['nombre'] == pokemon2, [f'debilidad_{tipo}' for tipo in df.columns if tipo.startswith('debilidad_')]]

    daño = ataque

    # Aplicar fortalezas
    daño *= (2 ** fortalezas_pokemon2.values)

    # Aplicar debilidades
    daño /= (2 ** debilidades_pokemon2.values)

    return daño

# Ejemplo de uso
pokemon1 = 'Bulbasaur'
pokemon2 = 'Charizard'
ataque_pokemon1 = df.loc[df['nombre'] == pokemon1, 'ataque'].values[0]

daño = calcular_daño(pokemon1, pokemon2, ataque_pokemon1)
print(f'Daño de {pokemon1} a {pokemon2}: {daño}')
```
